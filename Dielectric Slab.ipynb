```python
#!/usr/bin/env python

from __future__ import division

import numpy as np
import scipy as sp
from matplotlib import pyplot as plt

import putil
import sys
import os

from scipy.optimize import brentq
from scipy.interpolate import interp1d
from scipy.signal import convolve

pi = sp.pi
sqrt = sp.emath.sqrt

colors = ['r','b','g','c']


# Polarization according to waveguide convention - TE = E out of screen
# Theta is the angle between the slab surface (z axis) and the guided ray (ie, the compliment of the angle of incidence)
# Note that this is the opposite of the resonator/FDFD naming convention

# All wavevectors are normalized (divided) by ko (ie k_cladding)
# All thicknesses are normalized by high-index wavelength (ie d is actually d*n/lambda_o)

R_TM = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_TM_extended.csv',delimiter=',',dtype=complex)
R_TM1 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_TM_mode_2.txt',delimiter=',',dtype=complex)
R_TM2 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_TM_mode_3.txt',delimiter=',',dtype=complex)
R_TM3 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_TM_mode_4.txt',delimiter=',',dtype=complex)

R_ds = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_ds_extended.txt')
R_ds1 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_ds_mode_2.txt')
R_ds2 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_ds_mode_3.txt')
R_ds3 = np.loadtxt('C:\\Users\\Patrick\\Documents\\PhD\\Data\\Rectangular Resonator\\reflection_coefficient_ds_mode_4.txt')

def dispersion_TM(beta,ncore,nclad,d,m):
    '''
    from Inan Electromagnetic Waves pg 291, eq 4.40
    '''
    
    k = ncore    #k = ko*n, normalized by ko
    bx = sqrt(k**2 - beta**2)
    
    # a is transverse wavevector
    a = lambda b: sqrt((ncore**2-nclad**2)-b**2)
    
    return a(bx)/bx, (nclad/ncore)**2 * sp.tan(2*pi/ncore * bx*d/2. - m*pi/2.), a(bx)/bx - (nclad/ncore)**2 * sp.tan(2*pi/ncore * bx*d/2. - m*pi/2.)    #the first factor of 2pi/n in the tan() is to compensate for normalization


def numModes(ncore,nclad,d):
    critical_angle = sp.arcsin(nclad/ncore)
    M = sp.sin(pi/2. - critical_angle) / (1/(2*d))
        
    return int(np.ceil(M))
    
def numTotalModes(ncore,nclad,d):
    
    M = 2*d
        
    return int(np.floor(M))
    
        
def phase(d,pol='TM',mode=0):
    '''
    returns reflection phase as a function of waveguide thickness (units of high index wavelength)
    
    Computed using Junghyun Park's Matlab RCWA code
    '''
    
    if mode==0:
        if pol=='TM':
            angle = np.angle(R_TM,deg=True)
        else:
            angle = np.angle(R_TE,deg=True)
        
        interp = interp1d(R_ds,angle,kind='cubic')

    elif mode==1:
        if pol=='TM':
            angle = np.angle(R_TM1,deg=True)
        else:
            angle = np.angle(R_TE1,deg=True)
        
        interp = interp1d(R_ds1,angle,kind='linear')
            
    elif mode==2:
        if pol=='TM':
            angle = np.angle(R_TM2,deg=True)
        else:
            angle = np.angle(R_TE2,deg=True)
        
        interp = interp1d(R_ds2,angle,kind='linear')
            
    elif mode==3:
        if pol=='TM':
            angle = np.angle(R_TM3,deg=True)
        else:
            angle = np.angle(R_TE3,deg=True)
        
        interp = interp1d(R_ds3,angle,kind='linear')
            
    '''
    xs = np.linspace(0.02,3,1000)
    ys = interp(xs)

    plt.plot(ds,angle,'ro')
    plt.plot(xs,ys,'k-')
    plt.show()
    '''
    
    return interp(d)
    
def example_phase():
    xs = np.linspace(0.02,10,1000)
    ys = phase(xs,mode=3)

    plt.plot(xs,ys,'b-',lw=2)
    plt.title('Reflection Phase')
    plt.xlabel(r'Height/$\lambda$')
    plt.ylabel('Angle(R) (Degrees)')
    plt.show()

    
def Ref(d,pol='TM',mode=0):
    '''
    returns reflection magnitude as a function of waveguide thickness (units of high index wavelength)
    
    Computed using Junghyun Park's Matlab RCWA code
    '''
    if mode==0:
        if pol=='TM':
            mag = abs(R_TM)
        else:
            mag = abs(R_TE)
        
        interp = interp1d(R_ds,mag,kind='cubic')
        
    elif mode==1:
        if pol=='TM':
            mag = abs(R_TM1)
        else:
            mag = abs(R_TE1)
        
        interp = interp1d(R_ds1,mag,kind='cubic')
        
    elif mode==2:
        if pol=='TM':
            mag = abs(R_TM2)
        else:
            mag = abs(R_TE2)
        
        interp = interp1d(R_ds2,mag,kind='cubic')
        
    elif mode==3:
        if pol=='TM':
            mag = abs(R_TM3)
        else:
            mag = abs(R_TE3)
        
        interp = interp1d(R_ds3,mag,kind='cubic')
    

    '''
    xs = np.linspace(0.02,3,1000)
    ys = interp(xs)

    plt.plot(ds,angle,'ro')
    plt.plot(xs,ys,'k-')
    plt.show()
    '''
    
    return interp(d)
    
def example_Ref():
    xs = np.linspace(0.02,10,1000)
    ys = Ref(xs,mode=3)
    
    planar_r = (4-1)/(4+1)
    
    plt.plot(xs,ys,'r-',lw=2)
    plt.axhline(planar_r, color='k',linestyle='--')
    plt.ylim(0,1)
    plt.title('Reflection Magnitude')
    plt.xlabel(r'Height/$\lambda$')
    plt.ylabel('|R|')
    plt.show()
    
def Finesse(d,pol='TM',m=1):
    
    R = Ref(d,pol=pol,mode=m)
    
    return pi/(2*sp.arcsin((1-R)/(2*R**0.5)))

def Beta(ncore,nclad,d,m,pol='TE'):
    '''
    Returns propagation constant of the mth mode in a dielectric slab waveguide, thickness d in units of wls.
    '''
    
    critical_angle = sp.arcsin(nclad/ncore)
    
    if m/(2*d) > sp.sin(pi/2. - critical_angle): #leaky modes
        print 'Warning: Leaky mode at m =',m    
        if pol=='TE':
            theta = sp.arcsin(m/(2*d))
        else:
            theta = sp.arcsin(m/(2*d)) #needs to be corrected to include brewster's angle behavior
    
    elif m/(2*d) <= sp.sin(pi/2. - critical_angle): #bound modes
        
        if pol=='TE':
            F = lambda x: sp.tan(pi*d*x - m*pi/2.) - sp.sqrt(sp.sin(pi/2.-critical_angle)**2 / (x**2) - 1)
        else:
            F = lambda x: sp.tan(pi*d*x - m*pi/2.) - 1/sp.cos(pi/2.-critical_angle)**2 * sp.sqrt(sp.sin(pi/2.-critical_angle)**2 / (x**2) - 1)
        
        x0 = m/(2*d)
        x1 = (m+1)/(2*d)-1e-12
        
        '''
        xs = np.linspace(x0,x1,1000)
        ys = map(F,xs)
        plt.plot(xs,ys)
        plt.ylim(-5,5)
        plt.show()
        exit()
        '''
        
        sinth = brentq(F,x0,x1)
        
        theta = sp.arcsin(sinth)
        
    beta = ncore*2*pi*sp.cos(theta)
        
    return beta
    
def getBetas(ncore,nclad,d,pol='TE',mmax=None):
    '''
    Returns list of propagation constants for a dielectric slab waveguide, thickness d in units of wls (TE modes).
    '''
    #print d
    
    M = numTotalModes(ncore,nclad,d)
    
    Bs = []
    
    if mmax==None: l=M
    else: l=mmax
    
    for m in range(l):
        if m>M:
            Bs.append(np.nan)
        else:
            Bs.append(Beta(ncore,nclad,d,m,pol))
        
    return np.array(Bs)
    
def neff(B):
    '''
    converts values of Beta into mode indices
    '''
    
    wl = 2*sp.pi/B
    n = 1/wl
    return n

def example_neff():
    ds = np.linspace(0.02,3,100)
    ns_TE = np.zeros(len(ds))
    ns_TM = np.zeros(len(ds))
    
    for i,di in enumerate(ds):
        n = neff(getBetas(4,1,di,pol='TM')[0])
        n2 = neff(getBetas(4,1,di,pol='TE')[0])
        ns_TM[i] = n
        ns_TE[i] = n2
        
    plt.plot(ds,ns_TM,'r-',linewidth=2)
    plt.plot(ds,ns_TE,'b-',linewidth=2)
    plt.axhline(3.88,color='k',linestyle='--',linewidth=2)
    plt.xlabel(r'Height / $\lambda$')
    plt.ylabel(r'$n_{eff}$')
    plt.title('Mode Index')
    plt.legend(('TM','TE'),loc='lower right')
    plt.show()


def rect_cavity_modes():
    
    pol='TM'
    ncore = 4
    nclad = 1
    
    d = np.linspace(0.02,3,500) # in units of high-index wavelength
    mmax=4 # highest order of waveguide mode to consider (m = # of field antinodes)
    wmodes=7 # number of lateral resonances to consider (with no phase, the 1st will be a line at width=0)
    
    # Beta values of each waveguide mode for each slab thickness.  0th index is slab height, 1st index is mode.
    BB = np.zeros((len(d),mmax))
    
    for i,di in enumerate(d):
        Bs = getBetas(ncore,nclad,di,pol=pol,mmax=mmax)
        Bs /= ncore # convert from free-space wavelengths to high-index wavelengths
        
        BB[i,:]=Bs
    
    # w is a 3D matrix containing the resonant width for each slab thickness d, lateral mode, and waveguide mode order.  Indices are in the respective order.
    w = np.zeros((len(d),wmodes,mmax))
    
    for m in range(mmax):
        phi = phase(d,pol=pol,mode=m)*pi/180. 
        for i in range(wmodes):
            w[:,i,m] = (i*pi-phi)/BB[:,m]
    
    #line_plot(d,w)
    colormap_plot(d,w,pol=pol,mode='finesse')
    
    #plt.show()
    
    #return d,BB
    
def line_plot(d,w):
    
    plt.figure(figsize=(5,5))
    for i in range(len(w[0,:,0])):
        for j in range(len(w[0,0,:])):
            plt.plot(w[:,i,j],d,color=colors[j])
    
    plt.xlim(0.02,3)
    plt.ylim(0.02,3)
    plt.xlabel(r'Width / $\lambda$')    
    plt.ylabel(r'Height / $\lambda$')    
    plt.title('Truncated Dielectric Slab Resonances')

    return 0

def colormap_plot(d,w,pol='TM',mode='Q'):
    
    cmap = np.zeros((len(d),len(d)))
    
    
    
    for j in range(len(w[0,0,:])):
        for i in range(1,len(w[0,:,0])):
            
            #vert = np.zeros((len(d),len(d))) # matrix for vertically traveling resonances
            #horiz = np.zeros((len(d),len(d))) # matrix for horizontally traveling resonances

            # convert width values to index locations in d
            line = w[:,i,j]/d[-1]*len(d)
            line = np.rint(line).astype(int)
            line = line * (0 < line) * (line < len(d))

            di = np.arange(len(d))
            
            '''
            print w[:,i,j]
            print d[-1]
            print line
            print di
            '''
            
            if mode=='Q':
                cmap[line,di] = Finesse(d,pol=pol,m=j)*2*w[:,i,j]
                cmap[di,line] = Finesse(d,pol=pol,m=j)*2*w[:,i,j]
            elif mode=='finesse':
                cmap[line,di] = Finesse(d,pol=pol,m=j)
                cmap[di,line] = Finesse(d,pol=pol,m=j)
            
            #cmap += vert + horiz
            
        plt.figure(figsize=(5,5))
        
        # blur lines for aesthetic purposes
        cmap_blur = blur_image(cmap,5)
                
        plt.imshow(cmap_blur,extent=(d[0],d[-1],d[0],d[-1]),vmin=0,vmax=None, origin='lower')
        plt.colorbar()
        plt.xlabel(r'Width / $\lambda$')    
        plt.ylabel(r'Height / $\lambda$')    
        if mode=='Q': plt.title('Fabry-Perot Quality Factor')
        elif mode=='finesse': plt.title('Fabry-Perot Finesse')

    return 0

def example_dispersion():
    
    ncore = 4
    nclad = 1
    
    d = 0.2*ncore
    
    kmax = ncore
    re = np.linspace(0,1,1000)
    im = np.linspace(0,kmax*2,1000)
    
    B_re, B_im = np.meshgrid(re,im)

    B = B_re + 1j*B_im
    
    disp_odd = abs(dispersion_TM(B,ncore,nclad,d,0)[2])
    disp_even = abs(dispersion_TM(B,ncore,nclad,d,1)[2])
    disp = disp_odd*disp_even
    
    plt.imshow(sp.log10(disp),origin='lower',extent=[0,kmax,-kmax/2,kmax/2],vmax=2,cmap='jet',interpolation='nearest')
    plt.colorbar()
    plt.show()
    
def debug():
    
    ncore = sqrt(2)
    nclad = 1
    
    d = 1.25*ncore
    
    B = sp.linspace(0,2,100)
    
    ab,t,x = dispersion_TM(B,ncore,nclad,d,0)

    plt.plot(B,ab,'r-')
    plt.plot(B,t,'b-')
    plt.ylim(-1,1)
    plt.show()

def example_dispersion_Inan():
    ncore = sp.sqrt(2)
    nclad = 1
    
    d = 1.25
    
    bx = np.linspace(0,2*pi*2/d,500)
    a = lambda b: sp.sqrt((2*pi)**2*(ncore**2-nclad**2)-b**2)
    
    ax = np.array(map(a,bx))
    t = nclad**2/ncore**2 * sp.tan(bx*d/2)
    t2 = nclad**2/ncore**2 * sp.tan(bx*d/2+pi/2.)
    
    plt.figure(figsize=(5,5))
    plt.plot(bx*d/2./pi,ax*d/2./pi,'r-')
    plt.plot(bx*d/2./pi,t/pi,'b-')
    plt.plot(bx*d/2./pi,t2/pi,'b-')
    
    for i in range(4):
        plt.axvline(1/2*i,color='k',linestyle='--')
        
    plt.xlim(0,2)
    plt.ylim(0,2)
    plt.show()
    
def gauss_kern(size, sizey=None):
    """ Returns a normalized 2D gauss kernel array for convolutions """
    size = int(size)
    if not sizey:
        sizey = size
    else:
        sizey = int(sizey)
    x, y = np.mgrid[-size:size+1, -sizey:sizey+1]
    g = sp.exp(-(x**2/float(size)+y**2/float(sizey)))
    return g / g.sum()

def blur_image(im, n, ny=None) :
    """ blurs the image by convolving with a gaussian kernel of typical
        size n. The optional keyword argument ny allows for a different
        size in the y direction.
    """
    g = gauss_kern(n, sizey=ny)
    improc = convolve(im,g, mode='valid')
    return(improc)
    
```

It certainly doesn't appear like there's any alternating trend.  Any alternation should be present in the raw reflectance data from RCWA, I would think:

```python
xs = np.linspace(0.02,10,1000)

for i in (np.arange(4)):
    ys = Ref(xs,mode=i)
    plt.plot(xs,ys,lw=2)
    
plt.title('Reflection Amplitude')
plt.legend(('mode order = 1','mode order = 2','mode order = 3','mode order = 4'),loc='best')
plt.xlabel(r'Height/$\lambda$')
plt.ylabel('|R|')
plt.ylim(0,1)
```

Aha. My intuition was right - the symmetry of the modes prevents scattering into plane waves _in the direction of waveguide propagation_.  Just as coupling into the mode from a normal incidence plane wave is forbidden, but oblique angles is allowed, the truncated waveguide will still scatter strongly at non-normal angles.  Thus, the reflecatance of the truncated waveguide is not strongly affected by the symmetry of the mode.

getQ creates the Q matrix in Goell's paper; this matrix represents the system of equations for matching all field components at the boundary of the resonator.  The value returned by detQ, as one might expect, is the determinant of Q.  When this value is zero, the present set of dimensions, index, propagation constant, etc, correspond to an allowed mode of the system.

```python
def getQ(P,dims,nr,sym=1,N=5,harm=1,scaling=True):
    
    '''
    Matches amplitudes of cylindrical harmonics at boundary of rectangular dielectric region
    
    Inputs:
    P    - normalized propagation constant (range 0-1)
    dims - dimensions of resonator (width,height)
    nr   - refractive index of resonator
    sym  - Consider even (0) or odd (1) symmetry across x-axis
    N    - number of harmonics to consider
    harm - {0,1,'both') determines if even (0), odd (1) or both harmonics are considered in the cylindrical harmonic expansion
    '''
    
    kz = csqrt(P**2*(nr**2-1)+1)*ko
    h = sqrt(k1**2-kz**2)
    p = sqrt(kz**2-ko**2)
    
    a,b = dims
        
    if sym == 1:   # odd  mode w.r.t x-axis
        phi = 0
        
    elif sym == 0: # even mode w.r.t x-axis
        phi = pi/2.
    
    # Instantize matrices
    
    eLA = np.zeros((N,N),dtype=float) # N: number of harmonics we are using
    eLC = np.zeros((N,N),dtype=float)
    hLB = np.zeros((N,N),dtype=float)
    hLD = np.zeros((N,N),dtype=float)
    eTA = np.zeros((N,N),dtype=float)
    eTB = np.zeros((N,N),dtype=float)
    eTC = np.zeros((N,N),dtype=float)
    eTD = np.zeros((N,N),dtype=float)
    hTA = np.zeros((N,N),dtype=float)
    hTB = np.zeros((N,N),dtype=float)
    hTC = np.zeros((N,N),dtype=float)
    hTD = np.zeros((N,N),dtype=float)
    
    for ni in range(N): # arrat (0 to N-1)
        
        # angles used for boundary matching fields at boundary depend on whether current harmonic is odd/even 
        
        # use exclusively even or odd harmonics
        if harm==1: n=2*ni+1
        elif harm==0: n=2*ni
        else: n=ni
        
        m = np.arange(N) + 1  # m is 1 to N
        theta = (m-0.5)*pi/(2*N)  # theta_m
    
        # Formulate Matrix Elements
        
        tc = sp.arctan(b/a)

        R = sin(theta)         * (theta < tc) + cos(theta + pi/4.)  * (theta==tc) + -1*cos(theta)     * (theta > tc)
        T = cos(theta)         * (theta < tc) + cos(theta - pi/4.)  * (theta==tc) +    sin(theta)     * (theta > tc)
        rm = a/(2.*cos(theta)) * (theta < tc) + (a**2+b**2)**0.5/2. * (theta==tc) + b/(2.*sin(theta)) * (theta > tc)
                
        S = sin(n*theta + phi)
        C = cos(n*theta + phi)
        
        J = jn(n,h*rm)
        Jp = jvp(n,h*rm)
        JJ = n*J/(h**2*rm)
        JJp = Jp/(h)

        K = kn(n,p*rm)
        Kp = kvp(n,p*rm)
        KK = n*K/(p**2*rm)
        KKp = Kp/(p)
        
        # scaling to prevent overflow/underflow
        
        if scaling:
            d = (a+b)/2.
            Jmult = h**2*d/abs(jn(n,h*b))
            Kmult = p**2*d/abs(kn(n,p*b))
        else:
            Jmult = Kmult = 1.
        
        eLA[:,ni] = J*S                                          * Jmult
        eLC[:,ni] = K*S                                          * Kmult
        hLB[:,ni] = J*C                                          * Jmult
        hLD[:,ni] = K*C                                          * Kmult
        eTA[:,ni] = -1*kz*(JJp*S*R + JJ*C*T)                     * Jmult
        eTB[:,ni] = ko*Zo*(JJ*S*R + JJp*C*T)                     * Jmult
        eTC[:,ni] = kz*(KKp*S*R + KK*C*T)                        * Kmult
        eTD[:,ni] = -1*ko*Zo*(KK*S*R + KKp*C*T)                  * Kmult
        hTA[:,ni] = er*ko*(JJ*C*R - JJp*S*T)/Zo                  * Jmult                 # typo in paper - entered as JJp rather than Jp
        hTB[:,ni] = -kz*(JJp*C*R - JJ*S*T)                       * Jmult
        hTC[:,ni] = -ko*(KK*C*R - KKp*S*T)/Zo                    * Kmult
        hTD[:,ni] = kz*(KKp*C*R - KK*S*T)                        * Kmult
        
    O = np.zeros(np.shape(eLA))
    
    Q1 = np.hstack((eLA,O,-1*eLC,O))
    Q2 = np.hstack((O,hLB,O,-1*hLD))
    Q3 = np.hstack((eTA,eTB,-1*eTC,-1*eTD))
    Q4 = np.hstack((hTA,hTB,-1*hTC,-1*hTD))

    Q = np.vstack((Q1,Q2,Q3,Q4))
    
    # for even harmonics, eliminate n=0 terms for E or H, depending on symmetry.  Inclusion of these terms results in zero columns and thus a zero determinant.
    # Since we are eliminating columns from our matrix, we must eliminate rows as well to maintain square dimensions (4N-2).  Goell's
    # convention is to discard the first and last rows for whichever longitudinal component has odd symmetry (eg. hLB/D if sym=0)
    if harm==0:
        if sym==0:                     #eliminate b0,d0 terms
            Q = np.delete(Q,[N,3*N],1) #delete syntax: (array,index,axis (0 = row, 1 = column)
            Q = np.delete(Q,[N,2*N-1],0)
        elif sym==1:
            Q = np.delete(Q,[0,2*N],1)
            Q = np.delete(Q,[0,N-1],0)

        
    return Q
    
```

The values of the determinants for Q tend to be obnoxiously small (eg. ~1e-60).  However, near cutoff, there can be a lot of small-signal fluctuations in detQ, requiring some kind of cutoff value.  For now this has to be controlled manually.

```python
def detQ(P,dims,nr,sym=1,N=5,harm=1, thresh=1e70):
    dq = np.linalg.det(getQ(P,dims,nr,sym,N,harm))
    if abs(dq) > thresh: return dq
    else: return 0
```

Now that we have obtained the propagation constants of each mode, we can generate any pertinient information about the mode that we might be interested in.  Notably, we can compute each of the field components.  Below, we will determine the longitudinal E field components as a demonstration.  First, we prepare a grid of R and $\theta$ values for plotting the fields.  

```python
# resonator dimensions
b = 1; a = AR

res = 200

xmax = a + 1
ymax = b + 1

xs = np.linspace(-xmax,xmax,res)
ys = np.linspace(-ymax,ymax,res)

X,Y = np.meshgrid(xs,ys)

R = sqrt(X**2 + Y**2)
TH = sp.arctan(Y/X)

mask = (abs(X) <= a/2.) * (abs(Y) <= b/2.)
```

```python
def getphi(sym):
    if sym == 1:   # odd  mode w.r.t x-axis
        phi = 0
        
    elif sym == 0: # even mode w.r.t x-axis
        phi = pi/2.
        
    return phi

def getns(N,harm):
    n = np.arange(N)
    if harm==1: return 2*n+1
    else: return 2*n
```

At this point, we need to obtain the propagation constant for our mode of interest.  For simplicity we pick an arbitrary beta and find the mode with the largest propagation constant at that beta.  Obviously, this can be designed to choose whichever mode out of the array you would want.  Or, by specifying exact dimensions and wavelength, etc, you could calculate the specific P you are interested in.

```python
iB = 10
Ps = modes[:,:,iB]
Bi = BB[iB]
print Bi

P_fund = np.amax(Ps)
print P_fund

# we know the largest P will be associated with the fundamental mode, which (based on the color above) is taken from dqe1.
# We need to define sym and harm for this mode for the following code.

sym = 0
harm = 1
```

Calculate some useful numbers...

```python
wl = 2*b/Bi*sqrt(nr**2-1)

ko = 2*pi/wl
k1 = ko * nr

ns = getns(N,harm)
phi = getphi(sym)
```

We now build Q for the chosen propagation constant and solve Q*T = 0 by setting the first element of T = 1.  The matrix equation can then be rearranged into the form Ax=B and solved for x

```python
q = getQ(P_fund,(a,b),nr,sym,N,harm,scaling=False)

A = q[1:,1:]; B = -1*q[1:,0]

x = np.linalg.solve(A,B)
print np.amax(abs(np.dot(A,x) - B))
```

x now holds the a,b,c,d coefficients, except for whichever one we arbitrarily set to 1, and the a0,c0 etc terms if we were using even harmonics.  We reinsert those and then break the coefficient list into a,b,c,d lists

```python
coefs = np.insert(x,0,1)

if harm == 0:
    if sym == 0:
        coefs = np.insert(coefs,N,0)
        coefs = np.insert(coefs,3*N,0)
    elif sym == 1:
        coefs = np.insert(coefs,0,0)
        coefs = np.insert(coefs,2*N,0)
```

```python
A,B,C,D = np.split(coefs,4)
A,B,C,D
```

Coefficients and bessel functions in hand, we can now construct our fields 

```python
kz = Kz(P_fund)
h = csqrt(k1**2-kz**2)
p = csqrt(kz**2-ko**2)
Zo = 1

Ez_int = np.sum([A[ni]*jn(n,h*R)*sin(n*TH + phi) for ni,n in enumerate(ns)],axis=0)
Ez_ext = np.sum([C[ni]*kn(n,p*R)*sin(n*TH + phi) for ni,n in enumerate(ns)],axis=0)

Er_int = np.sum([ 1j*kz/h**2*(A[ni]*h*jvp(n,h*R)*sin(n*TH + phi) - ko*Zo/(kz*R)*B[ni]*jn(n,h*R)*sin(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)
Er_ext = np.sum([-1j*kz/p**2*(C[ni]*p*kvp(n,p*R)*sin(n*TH + phi) - ko*Zo/(kz*R)*D[ni]*kn(n,h*R)*sin(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)

Et_int = np.sum([ 1j*kz/h**2*(A[ni]/R*jn(n,h*R)*cos(n*TH + phi)*n - ko*Zo/(kz)*B[ni]*h*jvp(n,h*R)*cos(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)
Et_ext = np.sum([-1j*kz/p**2*(C[ni]/R*kn(n,p*R)*cos(n*TH + phi)*n - ko*Zo/(kz)*D[ni]*p*kvp(n,p*R)*cos(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)

Hz_int = np.sum([B[ni]*jn(n,h*R)*cos(n*TH + phi) for ni,n in enumerate(ns)],axis=0)
Hz_ext = np.sum([D[ni]*kn(n,p*R)*cos(n*TH + phi) for ni,n in enumerate(ns)],axis=0)

Hr_int = np.sum([ 1j*kz/h**2*(-1*(k1**2/(Zo*kz*R))*A[ni]*jn(n,h*R)*cos(n*TH + phi)*n + B[ni]*h*jvp(n,h*R)*cos(n*TH + phi)) for ni,n in enumerate(ns)],axis=0)
Hr_ext = np.sum([-1j*kz/p**2*(-1*(ko**2/(Zo*kz*R))*C[ni]*kn(n,p*R)*cos(n*TH + phi)*n + D[ni]*p*kvp(n,p*R)*cos(n*TH + phi)) for ni,n in enumerate(ns)],axis=0)

Ht_int = np.sum([ 1j*kz/h**2*(k1**2/(Zo*kz)*A[ni]*h*jvp(n,h*R)*sin(n*TH + phi) - B[ni]/R*jn(n,h*R)*sin(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)
Ht_ext = np.sum([-1j*kz/p**2*(ko**2/(Zo*kz)*C[ni]*p*kvp(n,p*R)*sin(n*TH + phi) - D[ni]/R*kn(n,p*R)*sin(n*TH + phi)*n) for ni,n in enumerate(ns)],axis=0)

Ez = Ez_int * mask + Ez_ext * (~mask)
Er = Er_int * mask + Er_ext * (~mask)
Et = Et_int * mask + Et_ext * (~mask)

Ey = Er*sin(TH) + Et*cos(TH)
Ex = Er*cos(TH) - Et*sin(TH)

Hz = Hz_int * mask + Hz_ext * (~mask)
Hr = Hr_int * mask + Hr_ext * (~mask)
Ht = Ht_int * mask + Ht_ext * (~mask)

Hy = Hr*sin(TH) + Ht*cos(TH)
Hx = Hr*cos(TH) - Ht*sin(TH)
```
